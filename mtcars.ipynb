```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

plt.rcParams['figure.figsize'] = (20,10)

df = pd.read_csv('mtcars.csv')
```

```python
df.head()
```

#### Membuat column `mpg_level` berdasarkan nilai `mpg`

```python
conditions = [
    (df['mpg'] < 20 ),
    (df['mpg'] > 30)]
choices = ['Low', 'Hard']
df['mpg_level'] = np.select(conditions, choices, default='Medium')
df
```

#### Bar plot jumlah mobil berdasarkan `mpg_level`

```python
df.groupby('mpg_level')['Cars'].count().plot(kind='bar')

plt.show()
```

#### Scatter plot dari `mpg` dan `qsec`

```python
df.plot.scatter(x='mpg', y='qsec')
```

```python
sns.set_context('talk')

g = sns.pairplot(df, hue="mpg_level")
```

#### Membandingkan histogram `qsec` untuk masing-masing jenis `mpg_level`

```python
plt.hist(df[df["mpg_level"]=="Low"]["qsec"], alpha=0.6, label="Low")
plt.hist(df[df["mpg_level"]=="Medium"]["qsec"], alpha=0.6, label="Medium")
plt.hist(df[df["mpg_level"]=="Hard"]["qsec"], alpha=0.6, label="Hard")
plt.legend()
plt.show()
```

#### Bar plot jumlah mobil berdasarkan Transmisi

```python
df['am'] = np.where(df['am'] == 0, "Automatic", 'Manual')
df
```

```python
df.groupby('am')['Cars'].count().plot(kind='bar')

plt.ylabel("Jumlah Mobil")
plt.xlabel("Transimisi")
plt.show()
```
