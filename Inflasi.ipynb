```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

plt.rcParams['figure.figsize'] = (20,10)

df = pd.read_csv('inflasi.csv')
```

```python
df.head()
```

```python
df[['Bulan','Tahun']] = df.Month.str.split(" ",expand=True,)
df.drop(columns =["Month"], inplace = True) 
```

```python
df['Inflasi'] = df['Inflasi'].apply(lambda x: x.replace('%', '')).astype('float') / 100
```

```python
fig, ax = plt.subplots(figsize=(20,10))
sns.set_context('talk')

df.groupby('Tahun')['Inflasi'].mean().plot(kind='bar')

plt.show()
```

### Analisis

- Rata-rata inflasi terbesar pada tahun 2006
- Sejak tahun 2013 sampai 2019, rata-rata inflasi mengalami penurunan, kenaikan hanya terjadi ditahun 2017 namun tidak terlalu besar

```python
fig, ax = plt.subplots(figsize=(15,10))
sns.set_context('poster')

sns.barplot(data=df,
         y='Inflasi',
         x='Tahun',
         hue='Bulan')
```

```python
fig, ax = plt.subplots(figsize=(15,10))
a = df["Tahun"]==2019
a = a.cumsum()

sns.barplot(data=df,
         y='Inflasi',
         x=a,
         hue='Bulan')
```

### Analisis

- Antara tahun 2003 dan 2019, puncak inflasi terbesar pada bulan Nopember 2005
- Tahun 2019 cenderung stabil tidak ada kenaikan atau penurunan secara signifikan
